```python
# Import Module 
from bs4 import BeautifulSoup 
import requests 

# Website URL 
URL = 'https://www.instagram.com/gajodharsinghcool/'

# class list set 
class_list = set() 

# Page content from Website URL 
page = requests.get( URL ) 

# parse html content 
soup = BeautifulSoup( page.content , 'html.parser') 

# get all tags 
tags = {tag.name for tag in soup.find_all()} 

# iterate all tags 
for tag in tags: 

# find all element of tag 
    for i in soup.find_all( tag ): 

    # if tag has attribute of class 
        if i.has_attr( "class" ): 

            if len( i['class'] ) != 0: 
                class_list.add(" ".join( i['class'])) 

print( class_list ) 
```

```python
from TikTokApi import TikTokApi
import asyncio
import os

ms_token = os.environ.get("ms_token", None)  # set your own ms_token


async def get_hashtag_videos():
    async with TikTokApi() as api:
        await api.create_sessions(ms_tokens=[ms_token], num_sessions=1, sleep_after=3)
        tag = api.hashtag(name="funny")
        async for video in tag.videos(count=30):
            print(video)
            print(video.as_dict)


if __name__ == "__main__":
    asyncio.run(get_hashtag_videos())
```

```python
import requests
from bs4 import BeautifulSoup
followers = []
following = []
for i in l[:21]:
    url =  i
    response = requests.get(url)

    if response.status_code == 200:

        soup = BeautifulSoup(response.content, 'html.parser')
        
        if len(soup.find_all('strong', attrs = {'data-e2e':"followers-count"}))!= 0:
            followers.append(soup.find_all('strong', attrs = {'data-e2e':"followers-count"}))
        else:
            followers.append('<strong>nil</strong>')
        if len(soup.find_all('strong', attrs = {'data-e2e':"following-count"}))!= 0:
            following.append(soup.find_all('strong', attrs = {'data-e2e':"following-count"}))
        else:
            following.append('<strong>nil</strong>')
            
        
        
    else:
        followers.append('<strong>no-response</strong>')
        following.append('<strong>no-response</strong>')
```

```python
(followers)
```

```python
followers_count = []
following_count = []
for item1,item2 in zip(followers, following):
    soup1 = BeautifulSoup(str(item1), 'html.parser')
    soup2 = BeautifulSoup(str(item2), 'html.parser')
    
    strong_tags1 = soup1.find_all('strong')
    strong_tags2 = soup2.find_all('strong')
    
    for item1, item2 in zip(strong_tags1, strong_tags2):
        followers_count.append(item1.get_text())
        following_count.append(item2.get_text())
        
print(len(followers_count))
print(len(following_count))
```
