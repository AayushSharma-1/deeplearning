```python
dataset_url = "http://download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file('flower_photos', origin = dataset_url, cache_dir = '.', untar = True)
```

```python

data_dir
```

```python
import pathlib
data_dir = pathlib.Path(data_dir)
data_dir
```

```python
len(list(data_dir.glob('*/*.jpg')))
```

```python
roses = list(data_dir.glob('roses/*'))
```

```python
Image.open(str(roses[2]))
```

```python
tulips = list(data_dir.glob('tulips/*'))
```

```python
flowers_images_dict = {
    'roses': list(data_dir.glob('roses/*')),
    'daisy': list(data_dir.glob('daisy/*')),
    'dandelion': list(data_dir.glob('dandelion/*')),
    'sunflowers': list(data_dir.glob('sunflowers/*')),
    'tulips': list(data_dir.glob('tulips/*'))
}
```

```python
flowers_labels_dict = {
    'roses' : 0,
    'daisy' : 1,
    'dandelion' : 2,
    'sunflowers' : 3,
    'tulips' : 4
}
```

```python
img = cv2.imread(str(flowers_images_dict['roses'][0]))
img.shape
```

```python
x, y = [], []

for flower_name, images in flowers_images_dict.items():
    for image in images:
        img = cv2.imread(str(image))
        resized_img = cv2.resize(img, (224, 224))
        x.append(resized_img)
        y.append(flowers_labels_dict[flower_name])
```

```python
x[0]
```

```python
x = np.array(x)
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 0)
```

```python
x_train_s = X_train / 255
x_test_s = X_test / 255
```

```python

plt.imshow(x[0])
```

```python
predicted = m.predict(np.array([x[0], x[1], x[2]]))
predicted = np.argmax(predicted, axis = 1)
predicted
```

```python
feature_extractor_model = "https://www.kaggle.com/models/google/mobilenet-v2/frameworks/TensorFlow2/variations/tf2-preview-feature-vector/versions/4"
```

```python
pretrained_model_without_top_layer = hub.KerasLayer(
    feature_extractor_model, input_shape = (224,224,3), trainable =False)
```

```python
num_of_flowers = 5

model = tf.keras.Sequential([
  pretrained_model_without_top_layer,
  tf.keras.layers.Dense(num_of_flowers)
])

model.summary()
```

```python
model.compile(
  optimizer="adam",
  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
  metrics=['acc'])

model.fit(x_train_s, y_train, epochs=5)
```

```python
model.evaluate(x_test_s,y_test)
```
